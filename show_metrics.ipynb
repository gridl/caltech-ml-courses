```python
import pickle
import os

%pylab inline
```

```python
main_dir = 'metrics'
model_name = 'fc_try_2_lr_0.0005'
```

```python
with open(os.path.join(main_dir,'%s_losses.pkl' % model_name), 'rb') as f:
    losses = pickle.load(f)
```

```python
with open(os.path.join(main_dir,'%s_accuracies.pkl' % model_name), 'rb') as f:
    accuracies = pickle.load(f)
```

```python
for k, v in losses.items():
    pylab.plot(v, label=k)
    pylab.legend(loc='best', frameon=False)
pylab.title('losses');
```

```python
for k, v in accuracies.items():
    pylab.plot(v, label=k)
    pylab.legend(loc='best', frameon=False)
pylab.title('accuracies');
```
